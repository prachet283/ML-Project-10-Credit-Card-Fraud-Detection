```python
# extracting the compressed dataset
from zipfile import ZipFile
dataset = '/content/creditcardfraud.zip'

with ZipFile(dataset , 'r') as zip:
  zip.extractall()
  print("The Dataset is extracted")
```

Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
#checking the missing values
credit_card_data.isnull().sum()
```

```python
# distribution of legit and fraud count
credit_card_data['Class'].value_counts()
```

The dataset is imbalanced

0 --> Normal Transaction

1 --> Fraud Transaction

```python
# seperating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values
credit_card_data.groupby('Class').mean()
```

Under Sampling

Build a sample dataset consisting similar distribution of normal and fraud transaction

Number of Fraud Transaction = 492

```python
legit_sample = legit.sample(n=492)
```

Concatenating

```python
new_dataset = pd.concat([legit_sample,fraud], axis = 0)
```

```python
new_dataset.to_csv('new_dataset_creditcard.csv')
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
credit_card_data.groupby('Class').mean()
```

Splitting the data into Features and Targets

```python
X = new_dataset.drop(columns='Class',axis=1)
Y = new_dataset['Class']
```

```python
print(X)
print(Y)
```

```python
X_train , X_test , Y_train , Y_test = train_test_split(X,Y,test_size=0.1,stratify=Y,random_state=2)
```

```python
print(X.shape , X_train.shape , X_test.shape)
```

Model Training

```python
model = LogisticRegression()
```

```python
#training the model with data
model.fit(X_train , Y_train)
```

Model Evaluation

```python
# training data accuracy
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction , Y_train)
```

```python
print('Accuracy on training data:', training_data_accuracy)
```

```python
# testing data accuracy
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction , Y_test)
```

```python
print('Accuracy on training data:', testing_data_accuracy)
```
